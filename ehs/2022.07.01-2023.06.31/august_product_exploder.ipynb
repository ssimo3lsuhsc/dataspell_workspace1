```python
import numpy
import pandas
import re

august_prof_dev = pandas.read_csv("August_Professional_Development2023-03-13_12_31_18.csv", dtype=pandas.StringDtype())
august_prof_dev
```

```python
line_splitter = re.compile(r"\s*\(Amount:.*?\r?\n")
august_prof_dev["Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)"] = august_prof_dev["Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)"].apply(lambda products: line_splitter.split(products))
```

```python
august_prof_dev
```

```python
august_prof_dev = august_prof_dev.explode("Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)")
august_prof_dev
```

```python
august_prof_dev = august_prof_dev[~august_prof_dev["Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)"].str.startswith("Total:")]
august_prof_dev
```

```python
august_prof_dev["Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)"].unique()
```

```python
august_prof_dev = august_prof_dev.sort_values(by="Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)", key= lambda trainings:  trainings.apply(lambda training: [
"EHS Orientation and Brain Development (Morning Slot)",
"EHS Orientation and Brain Development (Afternoon Slot)",
"Wellness Tuesday (Morning Slot)",
"Wellness Tuesday (Afternoon Slot)",
"DAP and Curriculum (Morning Slot)",
"DAP and Curriculum (Afternoon Slot)",
"Screenings and Difficult Conversations (Morning Slot)",
"Screenings and Difficult Conversations (Afternoon Slot)"
    ].index(training)))
august_prof_dev
```

```python
august_prof_dev["Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)"] = august_prof_dev["Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)"].str.replace(re.compile(r" \((Morning|Afternoon) Slot\)$"), "", regex=True)
august_prof_dev
```

```python
august_prof_dev["Attended"] = august_prof_dev["Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)"].apply(lambda value: "Tuesday" if value == "Wellness Tuesday" else "No")
august_prof_dev
```

```python
august_prof_dev = august_prof_dev.rename({
    "First Name": "first_name",
    "Last Name": "last_name",
    "Work Email": "work_email",
    "Select your training days and slots (50 individuals per slot; booked-out slots will be disabled)": "training",
    "Attended": "attended",
    "Select your school:": "school",
    "": "Entry Index"
}, axis="columns")
august_prof_dev
```

```python
august_prof_dev["entry_index"] = august_prof_dev.index
august_prof_dev.reset_index(drop=True)
```
